# Image to text

```python
# outputs for finetuned model
i_ranks_ou = np.zeros(i2t_score_ou.shape[0])
for index,score in enumerate(i2t_score_ou):
    inds = np.argsort(score)[::-1]
    # Score
    rank = 1e20
    for i in img2txt_ou[str(index)]:
        tmp = np.where(inds == i)[0][0]
        if tmp < rank:
            rank = tmp
    i_ranks_ou[index] = rank
    
r1_txt_ou = np.where(i_ranks_ou < 1)[0]
r5_txt_ou = np.where(i_ranks_ou < 5)[0]
r10_txt_ou = np.where(i_ranks_ou < 10)[0]
print("performance of the fine-tuned model on OOD")
print(len(r1_txt_ou),len(r5_txt_ou),len(r10_txt_ou))

# outputs for finetuned model - iid
i_ranks_iu = np.zeros(i2t_score_iu.shape[0])
for index,score in enumerate(i2t_score_iu):
    inds = np.argsort(score)[::-1]
    # Score
    rank = 1e20
    for i in img2txt_iu[str(index)]:
        tmp = np.where(inds == i)[0][0]
        if tmp < rank:
            rank = tmp
    i_ranks_iu[index] = rank
    
r1_txt_iu = np.where(i_ranks_iu < 1)[0]
r5_txt_iu = np.where(i_ranks_iu < 5)[0]
r10_txt_iu = np.where(i_ranks_iu < 10)[0]
print("performance of the fine-tuned model on IID")
print(len(r1_txt_iu),len(r5_txt_iu),len(r10_txt_iu))

# outputs for zero-shot model
i_ranks_oz = np.zeros(i2t_score_oz.shape[0])
for index,score in enumerate(i2t_score_oz):
    inds = np.argsort(score)[::-1]
    # Score
    rank = 1e20
    for i in img2txt_oz[str(index)]:
        tmp = np.where(inds == i)[0][0]
        if tmp < rank:
            rank = tmp
    i_ranks_oz[index] = rank
    
r1_txt_oz = np.where(i_ranks_oz < 1)[0]
r5_txt_oz = np.where(i_ranks_oz < 5)[0]
r10_txt_oz = np.where(i_ranks_oz < 10)[0]
print("performance of the zero-shot model on OOD")
print(len(r1_txt_oz),len(r5_txt_oz),len(r10_txt_oz))


i_ranks_iz = np.zeros(i2t_score_iz.shape[0])
for index,score in enumerate(i2t_score_iz):
    inds = np.argsort(score)[::-1]
    # Score
    rank = 1e20
    for i in img2txt_iz[str(index)]:
        tmp = np.where(inds == i)[0][0]
        if tmp < rank:
            rank = tmp
    i_ranks_iz[index] = rank
    
r1_txt_iz = np.where(i_ranks_iz < 1)[0]
r5_txt_iz = np.where(i_ranks_iz < 5)[0]
r10_txt_iz = np.where(i_ranks_iz < 10)[0]
print("performance of the zero-shot model on IID")
print(len(r1_txt_iz),len(r5_txt_iz),len(r10_txt_iz))
```

### Image-to-text retrieval : OOD Image-text pairs that got benefitted by the fine-tuning

```python
for index in needed_idx_o[:5]:
    score = incorrect_10_oz[index]
    inds = np.argsort(score)[::-1]
    captions_z = [texts_oz[i] for i in inds[:10]]
    score_u = i2t_score_ou[images_ou.index(in_images_oz[index])]
    inds_u = np.argsort(score_u)[::-1]
    captions_u = [texts_ou[i] for i in inds_u[:10]]
    print("Ground truth:", in_texts_oz[index])
    display(Image(url="./../data/multimodal/" + in_images_oz[index]))

    print("zero-shot top 10 texts")
    for i in range(len(captions_z)):
        print(str(i)+ ". " + captions_z[i])
    print("------------------------------------")
    print("fine-tuned top 10 texts")
    for i in range(len(captions_u)):
        print(str(i)+ ". " + captions_u[i])
    print("++++++++++++++++++++++++++++++++++++")
```
